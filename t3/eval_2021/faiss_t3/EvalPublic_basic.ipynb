# T3 Benchmarks For The FAISS Baseline and GPU Hardware (Public Query Set, basic config)

```python
import sys
sys.path.append("../..") # make sure we can import t3eval.py from the toplevel 't3' directory
import t3eval
import importlib
importlib.reload(t3eval)
evaluator = t3eval.Evaluator("FAISS Baseline (basic)", # name of team/algo
                             "./public_basic.csv", # exported data from eval runs
                             "../../baseline2021.json", # baseline constants
                             "../../competition2021.json", # competition constants
                             system_cost=22021.90, # participants must submit and document their system cost
                             verbose=True, # recommended True for debugging
                             is_baseline=True
                            )
evaluator.eval_all()
evaluator.show_summary(savepath="baseline_basic_summary.png")
```

## Deep1B Analysis

```python
%matplotlib inline
evaluator.plot_recall("deep-1B", 
        zoom=[[1000,5000],[0.90, 0.99]], # provide zoom subplot extents
        tweak=[[2000,0],[-1200,0.01]], # adjust text annotation
        savepath="baseline_basic_deep-1B_recall.png"
) 
```
