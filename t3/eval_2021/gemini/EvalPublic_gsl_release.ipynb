# T3 Benchmarks For The Gemini Algorithm and Hardware (Public Query Set)

```python
import sys
sys.path.append("../..") # make sure we can import t3eval.py from the toplevel 't3' directory
import t3eval
import importlib
importlib.reload(t3eval)
evaluator = t3eval.Evaluator("Gemini (gsl_release)", # name of team/algo
                             "./public_gsl_release.csv", # exported data from eval runs
                             "../../competition2021.json", # competition constants
                             baseline_path="../../baseline2021.json", # baseline constants
                             system_cost=55726.26, # participants must submit and document their system cost
                             verbose=False, # recommended True for debugging
                             show_baseline_table=True
                            )
evaluator.eval_all()
evaluator.show_summary(savepath="gemini_summary.png")
```

## Deep1B Analysis

```python
%matplotlib inline
evaluator.plot_recall("deep-1B", 
        zoom=[[1900,2600],[0.97, 0.99]], # provide zoom subplot extents
        tweak=[[2000,0],[0,-0.009]], # adjust text annotation
        savepath="gemini_deep-1B_recall.png"
) 
```
